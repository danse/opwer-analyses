```r
d <- read.csv('data//flex.csv')
```

data is about hourly jobs, that's why it's called `flex`. the abundance of flexible jobs is what motivated me to invest some time on upwork

```r
median(d$candidates)
```

half of the job offers have less than 12 candidates. so there is a long tail of offers where your application can be noticed more, but are they good offers?

```r
median(d$averageRate)
```

half of the job offers are paid less than $3.6 per hour. so we could imagine that the jobs with very few candidates are the ones corresponding to clients paying a low average ... who would like to work for those clients? let's verify our idea with a visualisation

```r
plot(d$candidates, d$averageRate)
```

as it's visible above, there are many jobs corresponding to clients that pay a decent average, but they still don't receive many applications. it would be interesting to see those jobs and apply to them, right? that's what i did using the open source tool [opwer](https://github.com/danse/opwer/). it indeed pointed me to some interesting gigs and saved me a lot of time, but this is for another post to discuss. for now i just want to present you the tools i built, and get some feedback from a wider community. i'm not that interested in investing on upwork more, but it could be worthy if anybody other than me found any value in this. _fast alone, further together_
